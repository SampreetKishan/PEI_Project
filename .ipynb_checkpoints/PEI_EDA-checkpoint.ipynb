# PEI project exploratory data analysis 

## Import the necessary modules 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Markdown
import re
```

##  EDA on customer dataset

```python
# let's bring in the datasets into pandas dataframes

#Let's look at the customers file first
customers = pd.read_excel('input_datasets/Customer.xls')
print("A simple preview")
print(customers.head(5))
print("\n-------------------------------------------------------------------------------------\n"*2)
print("The structure of the dataframe:")
print(customers.info()) 
print("\n-------------------------------------------------------------------------------------\n"*2)
print("Let's see if the Customer_ID column has duplicates")
print("Number of duplicate Customer IDs:", customers['Customer_ID'].duplicated().sum())
print("\n-------------------------------------------------------------------------------------\n"*2)
print("Let's also see if there are any customers with the same first name and last name combo")

name_grouped = customers.groupby(['First', 'Last']).size()

# Check if any group has a count greater than 1
has_duplicates = (name_grouped > 1).any()

if has_duplicates:
    display(Markdown("**Duplicate combinations found**"))
else:
    display(Markdown("**No duplicate names found**"))
print("\n-------------------------------------------------------------------------------------\n"*2)

print("let's see if there are any special characters in the names of our customers")
def has_non_alpha(text):
  """Checks if a given text contains non-alphabetic characters."""
  return bool(re.search('[^a-zA-Z]', text))


customers['has_non_alpha'] = customers.apply(lambda row: has_non_alpha(row['First']) or has_non_alpha(row['Last']), axis=1)
print("Here are the customers with non-alphanumeric names:")
print(customers[customers['has_non_alpha'] == True])


print("\n-------------------------------------------------------------------------------------\n"*2)
print("Let's see how many countries there are and how many customers we have for each country") 
customer_grouped = customers.groupby('Country').size().to_frame(name='count')
plt.bar(customer_grouped.index, customer_grouped['count'])
plt.show()
print("\n-------------------------------------------------------------------------------------\n"*2)
print("let's see the distribution of the age of our customers")
#sns.kdeplot(data=customers, x='Age', fill=True)
#plt.xlabel('Age')
#plt.ylabel('Density')
#plt.title('Density Plot of Customer Ages')
#plt.show()

plt.hist(customers['Age'], bins=10, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Customer Ages')
plt.show()

print("\n-------------------------------------------------------------------------------------\n"*2)
print("let's see if there are any outliers in the age of our customers (just to make sure)")


plt.boxplot(customers['Age'])
plt.xlabel('Age')
plt.title('Box Plot of Customer Ages')
plt.show()


```

#### _Based on this, it looks like our customers dataset is somewhat clean EXCEPT that the names of some customers have non-alphanumeric characters_
#### _no null values_ 
#### _we could potentially use the Customer_ID column(250 customers) as the primary key for this table_
#### _the first name and last name combos are unique_
#### _and the ages are in an acceptable range of ~20 - 80._

##  EDA on orders dataset

```python
orders = pd.read_csv('input_datasets/Order.csv')
print("A simple preview")
print(orders.head(5))
print("\n-------------------------------------------------------------------------------------\n"*2)
print("The structure of the dataframe:")
print(orders.info()) 
print("\n-------------------------------------------------------------------------------------\n"*2)
print("Let's see if the Order_ID column has duplicates")
print("Number of duplicate Order IDs:", orders['Order_ID'].duplicated().sum())
print("\n-------------------------------------------------------------------------------------\n"*2)
print("Let's see how many different items there are and how many orders we have for each item") 
items_grouped = orders.groupby('Item').size().to_frame(name='count')
plt.bar(items_grouped.index, items_grouped['count'])
plt.xticks(rotation=90)
plt.show()
print("\n-------------------------------------------------------------------------------------\n"*2)
print("let's see the distribution of the amount from the orders")
plt.hist(orders['Amount'], bins=10, edgecolor='black')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.title('Distribution of order amounts')
plt.show()
print("\n-------------------------------------------------------------------------------------\n"*2)
print("let's see if there are any outliers in the amount (just to make sure)")
plt.boxplot(orders['Amount'])
plt.xlabel('Amount')
plt.title('Box Plot of order amounts')
plt.show()
print("\n-------------------------------------------------------------------------------------\n"*2)
print("let's see if all the values in Customer_ID column of the orders table map to a value in the Customer_ID column of the customers table")
if orders['Customer_ID'].isin(customers['Customer_ID']).all():
    display(Markdown("**All order customer IDs exist in the customers tables**"))
    
else:
    display(Markdown("**Some order customer IDs do not exist in customers**"))
    print(orders[~orders['Customer_ID'].isin(customers['Customer_ID'])])
```

#### _Looks like the Orders table is clean too_
#### _No null values_
#### _Order_ID is unique_
#### _Even though the boxplot might show there are some outliers, they are legitimate values. There is way too many items that are less than $2,000_
#### _Made sure that all the Customer_ID values in the orders table have a match with the Customer_ID column of the customers table._


## Let's look at shipping json file now 

```python
shipping = pd.read_json('input_datasets/Shipping.json')
```

```python
print(shipping.head(5))
print("\n-------------------------------------------------------------------------------------\n"*2)
print("The structure of the dataframe:")
print(shipping.info()) 
print("\n-------------------------------------------------------------------------------------\n"*2)
print("Let's see if the Shipping_ID column has duplicates")
print("Number of duplicate Shipping IDs:", shipping['Shipping_ID'].duplicated().sum())
print("\n-------------------------------------------------------------------------------------\n"*2)
print("Let's the unique Status values and the count of it")
status_counts = shipping['Status'].value_counts()
status_counts.plot(kind='bar')
plt.title('Count of Shipping Status')
plt.xlabel('Status')
plt.ylabel('Count')
plt.show()
print("\n-------------------------------------------------------------------------------------\n"*2)
print("let's see if the Customer_ID column of the shipping dataset has a unique match with the customer_ID column of the customers dataset")

if shipping['Customer_ID'].isin(customers['Customer_ID']).all():
    display(Markdown("**All shipping customer IDs exist in the customers tables**"))
    
else:
    display(Markdown("**Some shipping customer IDs do not exist in customers**"))
    print(orders[~shipping['Customer_ID'].isin(customers['Customer_ID'])])
```

#### _All in all the shipping dataset looks good by itself._
#### _No null values_
#### _all the shipping_id values are unique and can be used as primary key for the table._
#### _the Customer_ID column's values have a unique match to the Customer_ID column of the Customers table._
